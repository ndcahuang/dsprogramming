```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
lat_0 = data[data['shot_made_flag'] == 0]['lat']
lon_0 = data[data['shot_made_flag'] == 0]['lon']
```

```python
lat_1 = data[data['shot_made_flag'] == 1]['lat']
lon_1 = data[data['shot_made_flag'] == 1]['lon']
```

```python
f, ax = plt.subplots(1, 2, figsize=(15, 10), sharey=True)

ax[0].scatter(lon_0, lat_0, s=1, color = 'blue')
ax[0].set_title("Shot Missed in terms of [lon, lat]")
ax[0].set_xlabel("Longitude")
ax[0].set_ylabel("Latitude")

ax[1].scatter(lon_1, lat_1, s=1, color = 'red')
ax[1].set_title("Shot Made in terms of [lon, lat]")
ax[1].set_xlabel("Longitude")
ax[1].set_ylabel("Latitude")

plt.tight_layout()
```

```python
loc_x_0 = data[data['shot_made_flag'] == 0]['loc_x']
loc_y_0 = data[data['shot_made_flag'] == 0]['loc_y']
```

```python
loc_x_1 = data[data['shot_made_flag'] == 1]['loc_x']
loc_y_1 = data[data['shot_made_flag'] == 1]['loc_y']
```

```python
f, ax = plt.subplots(1, 2, figsize=(15, 10), sharey=True)
ax[0].scatter(loc_x_0, loc_y_0, s=1, color = 'blue')
ax[0].set_title("Shot Missed in terms of [loc_x, loc_y]")
ax[0].set_xlabel("Location X")
ax[0].set_ylabel("Location Y")

ax[1].scatter(loc_x_1, loc_y_1, s=1, color = 'red')
ax[1].set_title("Shot Made in terms of [loc_x, loc_y]")
ax[1].set_xlabel("Location X")
ax[1].set_ylabel("Location Y")

plt.tight_layout()
```

```python
f, ax = plt.subplots(1, 2, figsize=(15, 10))
ax[0].hist(loc_x_1)
ax[0].set_title("Distribution of Location X")
ax[0].set_xlabel("Location X")
ax[0].set_ylabel("Frequency")

ax[1].hist(loc_y_1, orientation = "horizontal")
ax[1].set_title("Distribution of Location Y")
ax[1].set_xlabel("Frequency")
ax[1].set_ylabel("Location Y")

plt.tight_layout()
```
